```python
def lerNotas():
  n = int(input("Informe a nota:"))
  return n

def resultado(n1,n2):
  media = (n1+n2)/2
  print("\r")
  print("Nota 1:",n1)
  print("Nota 2:",n2)
  print("\r")
  print("Média:", media, "\nResultado: ", end="")
  if media >= 6:
    print("Aprovado")
  else:
    print("Reprovado")


a = lerNotas()
b = lerNotas()
resultado(a,b)
```
